Metodi basati sulla decomposizione della matrice A e hanno complessitÃ  O(k n^2). 

Si considera la matrice A come somma di 3 matrici, A = D + E + F, con
 - D matrice diagonale 
 - E triangolare INFERIORE
 - F triangolare SUPERIORE 

Sono 3: 
 1. Metodo di Jacobi
 2. Metodo di Gauss-Seidel 
 3. Metodo di Gauss-Seidel SOR

```python
"""
A = M - N
M = D
N = -(E + F)
Tj = M(^-1) @ N = -D(^-1) @ (E + F)
x = (T @ x0) + (invM @ b)
""" 
def jacobi(A,b,x0,toll,it_max):
    errore=1000
    d = np.diag(A)
    n = A.shape[0]
    invM = np.diag(1/d)

    E = np.tril(A, -1) #to do: E triangolare inferiore
    F = np.triu(A, 1) #to do: F triangolare superiore
    N = -(E + F) #to do

    T = invM @ N #to do
    autovalori=np.linalg.eigvals(T)
    raggiospettrale = np.max(np.abs(autovalori)) #to do
    print("raggio spettrale jacobi", raggiospettrale)
    
    it=0 
    er_vet=[]
    while er_vet >= toll and it < it_max:#to do:
        x = (T @ x0) + (invM @ b) #to do
        errore=np.linalg.norm(x-x0)/np.linalg.norm(x)
        er_vet.append(errore)
        x0=x.copy()
        it=it+1
    return x,it,er_vet
```

```python
"""
Metodo iterativo per la risoluzione di sistemi lineari Ax = b, che si basa sulla decomposizione della matrice A
in tre matrici D (diagonale), E (triangolare inferiore), F (triangolare superiore). 
In generale, vale: 
    A = D + E + F, 
    A = M - N
    Tj = invM @ N

Per Gauss-Seidel vale: 
    M = E + D
    N = -F
"""
def gauss_seidel(A,b,x0,toll,it_max):
    errore=1000
    d = np.diag(A)

    D = np.diag(1/d) #to do
    E = np.tril(A, -1) #to do: E triangolare inferiore
    F = np.triu(A, 1) #to do: F triangolare superiore
    M = E + D #to do
    N = -F #to do
    T = npl.inv(M) @ N #to do

    autovalori=np.linalg.eigvals(T)
    raggiospettrale = np.max( np.abs(autovalori) ) #to do
    print("raggio spettrale Gauss-Seidel ",raggiospettrale)
    it=0
    er_vet=[]
    while er_vet >= toll and it < it_max: #to do:
        temp = b - F@x0 #to do
        x = SolveTriangular.Lsolve(M, temp) #to do 

        errore=np.linalg.norm(x-x0)/np.linalg.norm(x)
        er_vet.append(errore)
        x0=x.copy()
        it=it+1
        
    return x,it,er_vet
```
